```python
# Celda 1 — Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("✅ Librerías cargadas correctamente")
```

```python
# Celda 2 — Cargar el dataset
df = pd.read_csv('../../data/raw/customer_shopping_data.csv')

print(f"✅ Dataset cargado correctamente")
print(f"   Filas    : {df.shape[0]:,}")
print(f"   Columnas : {df.shape[1]}")
```

```python
# Celda 3 — Inspección inicial
print("=== INFORMACIÓN GENERAL ===")
df.info()
print()
print("=== PRIMERAS 5 FILAS ===")
df.head()
```

```python
# Celda 4 — Limpieza de datos

# 1. Convertir invoice_date de texto a fecha
df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True)

# 2. Crear columnas derivadas de la fecha (útiles para los módulos de temporada y eventos)
df['año']        = df['invoice_date'].dt.year
df['mes']        = df['invoice_date'].dt.month
df['dia_semana'] = df['invoice_date'].dt.day_name()

# 3. Crear columna de gasto total por transacción
df['total_compra'] = df['quantity'] * df['price']

# 4. Estandarizar texto
df['gender']        = df['gender'].str.strip().str.title()
df['category']      = df['category'].str.strip().str.title()
df['payment_method']= df['payment_method'].str.strip().str.title()
df['shopping_mall'] = df['shopping_mall'].str.strip().str.title()

print("=== DESPUÉS DE LA LIMPIEZA ===")
print(f"Filas          : {len(df):,}")
print(f"Columnas       : {df.shape[1]}")
print(f"Nulos          : {df.isnull().sum().sum()}")
print(f"Rango de fechas: {df['invoice_date'].min().date()} → {df['invoice_date'].max().date()}")
print(f"Total compras  : ${df['total_compra'].sum():,.0f}")
```

```python
# Celda 5 — Exportar dataset limpio
import os
os.makedirs('../../data/processed', exist_ok=True)

df.to_csv('../../data/processed/datos_limpios.csv', index=False)

print("✅ Dataset limpio guardado en data/processed/datos_limpios.csv")
print("   Este archivo se usará en todos los módulos del proyecto")
```
